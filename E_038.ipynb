Crear una Tupla Nombrada para Representar un Punto en el Plano

```python
# Importamos la clase namedtuple del módulo collections, y el módulo pyplot de la biblioteca matplotlib
from collections import namedtuple
import matplotlib.pyplot as plt

# Definimos una tupla nombrada llamada 'Punto' con campos 'x' e 'y'
Punto = namedtuple('Punto', ['x', 'y'])

# instancias de la tupla nombrada con valores específicos para x , y
coordenada_uno = Punto(5, 2)
coordenada_dos = Punto(-2, 3)
coordenada_tres = Punto(-1, -5)

# Extraemos los valores de las instancias de la tupla
x = coordenada_uno[0], coordenada_dos[0], coordenada_tres[0]
y = coordenada_uno[1], coordenada_dos[1], coordenada_tres[1]

# Etiquetas para los ejes x , y con un título
plt.xlabel('Coordenada X')
plt.ylabel('Coordenada Y')
plt.title('Puntos en el Plano')

# Creamos un gráfico de puntos
# Usando "o" para representar los puntos y markersize para especificar el tamaño
plt.plot(x, y, "o", markersize=10)

# Mostramos el gráfico generado
plt.show()



```
