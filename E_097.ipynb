**Análisis de datos de criptomonedas ficticias**

```python
import pandas as pd
import matplotlib.pyplot as plt
import random
from datetime import datetime, timedelta

# Función para generar datos ficticios de criptomonedas
def generar_datos(num_meses=12, num_criptomonedas=3):
  # Obtenemos la fecha actual
  fecha_actual = datetime.now()

  # Crea una lista de fechas retrocediendo en el tiempo
  fechas = [fecha_actual - timedelta(days=30 * i) for i in range(num_meses)][::-1]

  # Inicializar un diccionario con las fechas
  datos = {'Fecha': fechas}

  # Generar *precios aleatorios* para cada criptomoneda
  for i in range(num_criptomonedas):
    nombre = f'Cripto{i+1}'
    precios = [random.uniform(50, 200) for _ in range(num_meses)]
    datos[nombre] = precios

  # Crear un DataFrame con los datos generados
  df = pd.DataFrame(datos)
  return df

# Función para visualizar datos mensuales
def visualizar_datos_mensuales(df):
  # Configuramos el tamaño de la figura
  plt.figure(figsize=(12, 6))

  # Grafica los precios de cada criptomoneda en función del tiempo
  for columna in df.columns[1:]:
    plt.plot(df['Fecha'], df[columna], label=columna, marker='o', linestyle='-')

  # Configura título y etiquetas de ejes
  plt.title('Análisis Mensual de Criptomonedas Ficticias', fontsize=16)
  plt.xlabel('Fecha', fontsize=12)
  plt.ylabel('Precio', fontsize=12)

  # Mostramos leyenda y cuadrícula
  plt.legend()
  plt.grid(True)

  # Ajustamos el diseño y mostrar la gráfica
  plt.tight_layout()
  plt.show()

# Función para realizar análisis mensuales
def analisis_mensual(df):
  # Crear una nueva columna 'Mes' en el DataFrame
  df['Mes'] = df['Fecha'].dt.strftime('%Y-%m')

  # Calcular el promedio mensual para cada criptomoneda
  promedio_mensual = df.groupby('Mes').mean()

  # Imprimir el resultado del análisis mensual
  print("Promedio Mensual:")
  print(promedio_mensual)

# Generar datos ficticios mensuales
datos_cripto_mensuales = generar_datos()

# Visualizar datos mensuales
visualizar_datos_mensuales(datos_cripto_mensuales)

# Realizar análisis mensual
analisis_mensual(datos_cripto_mensuales)

```
