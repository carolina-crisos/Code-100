**Acceder a una URL e Imprimir su Contenido HTML**

```python
# Forma num. 1
# Importa la clase HTTPConnection del módulo http.client
from http.client import HTTPConnection

# URL del sitio web al que se realizará la conexión
url = 'google.com'

# Crea una instancia de la clase HTTPConnection utilizando la URL proporcionada
conexion = HTTPConnection(url)

# Envía una solicitud 'GET' al recurso raíz (página principal) '/' en el servidor web
conexion.request('GET', '/')

# Obtiene la respuesta del servidor
resultado = conexion.getresponse()

# Lee el contenido de la respuesta
contenido = resultado.read()

# Imprime el contenido obtenido de la respuesta del servidor
print(contenido)
```

```python
# Forma num. 2
# Importa el módulo 'request' desde la biblioteca 'urllib'
from urllib import request

# Realiza una solicitud a la URL proporcionada y obtiene los datos de la página web
pagina = request.urlopen("https://www.openwebinars.net")

# Lee los datos de la página web obtenidos mediante la solicitud
datos = pagina.read()

# Imprime los datos de la página web
print(datos)
```
