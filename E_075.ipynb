**imprimir todos los enlaces que existen en la web que hemos indicado.**

```python
# Importa el módulo urllib.request para realizar solicitudes web
import urllib.request

# Realiza una solicitud a la URL proporcionada y lee los datos de la página web
datos = urllib.request.urlopen('https://www.cyberclick.es/').read().decode()

# Importa la clase BeautifulSoup del módulo bs4 (Beautiful Soup) para el análisis HTML
from bs4 import BeautifulSoup

# Crea un objeto BeautifulSoup para analizar el contenido HTML de la página
soup = BeautifulSoup(datos)

# Encuentra todos los elementos 'a' (enlaces) en el contenido HTML
tags = soup('a')

# Itera sobre cada elemento 'a' encontrado
for tag in tags:
    # Imprime la URL (href) de cada enlace
    print(tag.get('href'))
```
