**Analiza datos utilizando Pandas y Matplotlib**

```python
# Importa las librerías necesarias
import pandas as pd
import matplotlib.pyplot as plt

# Datos de ventas en un diccionario
datos_ventas = {
    'Mes': ['Enero', 'Febrero', 'Marzo', 'Abril', 'Mayo'],
    'Ventas': [35000, 20000, 5000, 32000, 21000]
}

# Crea un DataFrame con los datos de ventas
df_ventas = pd.DataFrame(datos_ventas)

# Imprime los datos de ventas
print("Datos de ventas mensuales:")
print(df_ventas)

# Grafico de barras de las ventas mensuales
print()
plt.bar(df_ventas['Mes'], df_ventas['Ventas'], color='green')
plt.title('Ventas Mensuales')
plt.xlabel('Mes')
plt.ylabel('Ventas')
plt.show()

```
